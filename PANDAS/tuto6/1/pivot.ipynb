```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather.csv")
df
```

```python
#Pivot
#it allows u to transforme or reshape your data
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='date',columns='city',values='humidity')
```

```python
df.pivot(index='humidity',columns='city')
```

```python
#Pivot Table
#it allows u to summerize and aggreagate data inside dataframe
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")#that gives the average temperature
```

```python
#using aggregate functions
df.pivot_table(index="city",columns="date",aggfunc="sum")#sum
#default aggfunc is mean (average)
```

```python
#Margins
df.pivot_table(index="city",columns="date", margins=True)#also it uses mean as default aggfunc
```

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
```

```python
#Grouper

df = pd.read_csv("weather3.csv")
df
```

```python
#if you do not do that you will get an error cz date here is string so u should convert it first to date
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='ME',key='date'),columns='city')#M refers to month
```
