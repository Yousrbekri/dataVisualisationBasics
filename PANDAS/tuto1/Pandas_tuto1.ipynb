```python
#diffrent ways of creating dataframe
#
#
#
#df = pd.read_csv('HollywoodMovies.csv') # from csv

#df = pd.read_excel('HollywoodMovies.xlsx','Sheet1') #from excel

#from dictionnary
weather_data = {
    'day':['1/1/2017','1/1/2018','1/3/2017'],
    'temperature':[32,35,25],
    'event':['rain','sunny','snow']
}
df = pd.DataFrame(weather_data)
df
```

```python

#from tuple
weather_data = [
    ('1/1/2017',12,'rain'),
    ('1/1/2017',30,'sunny'),
    ('1/1/2017',5,'snow')
]
df = pd.DataFrame(weather_data,columns=['day','temperature','event'])
df
```

```python
#from list of dictionnary
weather_data = [
    {'day':'1/1/2017','temperature':12,'event':'rain'},
    {'day':'1/1/2018','temperature':25,'event':'sunny'},
    {'day':'1/3/2017','temperature':-5,'event':'snow'},
]
df = pd.DataFrame(weather_data)
df
```
