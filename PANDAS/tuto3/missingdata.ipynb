```python
#HANDLING MISSING DATA

import pandas as pd
df = pd.read_csv("weather_data.csv")
df
```

```python
import numpy as np
#methode to replace values in dataframe
#new_df = df.replace([-9999],np.nan)#Replacing single value
#replace per column  using dictionary
new_df = df.replace({
    'temperature':-9999,
    'windspeed':-9999,
    'event':'0'
   },np.nan)

new_df
```

```python
new_df = df.replace({#replacing by mapping
    -9999:np.nan,
    'No Event':'Sunny'
   })

new_df
```

```python
new_df = df.replace({  #replace data with regular expression
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
  },'',regex=True)

new_df
```

```python
#2

import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])#force the attrubut day to be date
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)#to change it in the original dataframe
df
```

```python
#FILLNA  Fill all NaN with one specific value
new_df = df.fillna(0)
new_df
```

```python
#Fill na using column names and dict
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
#Use method to determine how to fill na values
new_df = df.fillna(method="ffill") #replace NaN with the following value
new_df
```

```python
new_df = df.fillna(method="bfill")#replace NaN with the Previous value
new_df
```

```python
#Use of axis
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
#limit parameter
new_df = df.fillna(method="ffill",limit=1)#limit coppy
new_df
```

```python
#interpolate
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
```
